## Confusion Matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns

cm = metrics['confusion_matrix']

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## Phan bo Confidence cua Pseudo Labels

```python
import json
import pandas as pd

with open(DEFAULT_PSEUDO_LABEL_PATH, encoding='utf-8') as file:
    pseudo_labels = json.load(file)

pseudo_df = pd.DataFrame(pseudo_labels)
pseudo_df.head()
```

```python
if not pseudo_df.empty:
    plt.figure(figsize=(8, 5))
    sns.histplot(pseudo_df['confidence'], bins=20)
    plt.xlabel('Confidence')
    plt.ylabel('Number of samples')
    plt.title('Pseudo-label Confidence Distribution')
    plt.show()
else:
    print('Chua co pseudo label nao dat nguong confidence.')
```
